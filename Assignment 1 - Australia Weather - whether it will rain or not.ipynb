## Handling Missing Values

```python
#columns without null values
notnull_columns=weather.columns[weather.notnull().any()]
notnull_columns
```

```python
#finding out percentage of null values in a field
null_columns=weather.columns[weather.isnull().any()]
weather[null_columns].isnull().sum()/len(weather)*100
```

```python
#distribution of variables with high null values
low_quality_vals = ["Evaporation", "Sunshine", "Cloud9am", "Cloud3pm"]
for val in low_quality_vals:
    sns.distplot(weather[val])
    plt.show()
```

```python
#switching RainToday to Binary allows for correlation heatmap analysis later on
weather['RainToday'] = weather['RainToday'].str.replace('Yes', "1")
weather['RainToday'] = weather['RainToday'].str.replace('No', "0")
weather['RainToday'] = pd.to_numeric(weather['RainToday'], errors='coerce')
```

```python
#switching RainTomorrow to Binary allows for correlation heatmap analysis later on
weather['RainTomorrow'] = weather['RainTomorrow'].str.replace('Yes', "1")
weather['RainTomorrow'] = weather['RainTomorrow'].str.replace('No', "0")
weather['RainTomorrow'] = weather['RainTomorrow'].astype('int')
```

```python
lv, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(weather.corr(), annot=True, linewidths=0.5, linecolor = 'black', fmt= '.1f',ax=ax, cmap="coolwarm")
plt.show()
```

```python
#presenting correlation of variables with high number of null values
lv, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(weather[['Evaporation','Sunshine','Cloud9am','Cloud3pm','RainTomorrow']].corr(), annot=True, linewidths=0.5, linecolor = 'black', fmt= '.1f',ax=ax, cmap="coolwarm")
plt.show()
```

```python
#further check for distinctive patterns on variables vs whether it rains on next day
low_quality_vals = ["Evaporation", "Sunshine", "Cloud9am", "Cloud3pm"]
msk2 = weather['RainTomorrow'] == 0
msk3 = weather['RainTomorrow'] == 1
for val in low_quality_vals:
    msk1 = weather[val].notnull()
    target0 = weather[msk1 & msk2]
    target1 = weather[msk1 & msk3]
    sns.distplot(target0[val], hist=False, rug=True)
    sns.distplot(target1[val], hist=False, rug=True)
    plt.legend(labels=['Rain Tomorrow = No','Rain Tomorrow = Yes'])
    plt.xlabel(f'{val}')
    plt.title(f'Probability Distribution of {val} vs Rain Tomorrow')
    plt.show()
```

```python
#remade Sunshine graph from above for report
msk2 = weather['RainTomorrow'] == 0
msk3 = weather['RainTomorrow'] == 1
msk1 = weather["Sunshine"].notnull()
target0 = weather[msk1 & msk2]
target1 = weather[msk1 & msk3]
sns.distplot(target0["Sunshine"], hist=False, rug=True)
sns.distplot(target1["Sunshine"], hist=False, rug=True)
plt.legend(labels=['Rain Tomorrow = No','Rain Tomorrow = Yes'])
plt.xlabel('Hours of Bright Sunshine on the Day (Hr)')
plt.title('Probability Distribution of "Sunshine" vs Rain Tomorrow')
plt.show()
```

```python
#we see high correlation between humidity at 3pm vs whether it rains on the next day
msk2 = weather['RainTomorrow'] == 0
msk3 = weather['RainTomorrow'] == 1
msk1 = weather["Humidity3pm"].notnull()
target0 = weather[msk1 & msk2]
target1 = weather[msk1 & msk3]
sns.distplot(target0["Humidity3pm"], hist=False, rug=True)
sns.distplot(target1["Humidity3pm"], hist=False, rug=True)
plt.legend(labels=['Rain Tomorrow = No','Rain Tomorrow = Yes'])
plt.xlabel('Level of humidity at 3pm on the Day (%)')
plt.title('Probability Distribution of Humidity at 3pm vs Rain Tomorrow')
plt.show()
```

```python
#we see some correlation between humidity at 9am vs whether it rains on the next day
msk2 = weather['RainTomorrow'] == 0
msk3 = weather['RainTomorrow'] == 1
msk1 = weather["Humidity9am"].notnull()
target0 = weather[msk1 & msk2]
target1 = weather[msk1 & msk3]
sns.distplot(target0["Humidity9am"], hist=False, rug=True)
sns.distplot(target1["Humidity9am"], hist=False, rug=True)
plt.legend(labels=['Rain Tomorrow = No','Rain Tomorrow = Yes'])
plt.xlabel('Level of humidity at 9am on the Day (%)')
plt.title('Probability Distribution of Humidity at 9am vs Rain Tomorrow')
plt.show()
```

```python
#finding out percentage of null values in a field
null_columns=weather.columns[weather.isnull().any()]
weather[null_columns].isnull().sum()/len(weather)*100
```

```python
#We will drop datapoints for fields have less than 1% null values
weather.drop(columns = ['MinTemp','MaxTemp','Rainfall','WindSpeed9am','Temp9am', 'RainToday'], inplace = True)
```

```python
#replace NaN values with mean or mode or other categorical variables
#depending on the value as categorical or numerical 
weather["WindGustDir"].fillna("NA", inplace = True)     
weather["WindGustSpeed"].fillna(weather["WindGustSpeed"].mean(), inplace = True)  
weather["WindDir9am"].fillna("NA", inplace = True)      
weather["WindDir3pm"].fillna("NA", inplace = True)     
weather["WindSpeed3pm"].fillna("NA", inplace = True)
weather["Sunshine"].fillna(weather["Sunshine"].mean(), inplace = True)
weather["Cloud3pm"].fillna(weather["Cloud3pm"].mean(), inplace = True)
weather["Cloud9am"].fillna(weather["Cloud9am"].mean(), inplace = True)
weather["Humidity9am"].fillna(weather["Humidity9am"].mean(), inplace = True)
weather["Humidity3pm"].fillna(weather["Humidity3pm"].mean(), inplace = True)    
weather["Pressure9am"].fillna(weather["Pressure9am"].mean(), inplace = True)   
weather["Pressure3pm"].fillna(weather["Pressure3pm"].mean(), inplace = True) 
weather["Temp3pm"].fillna(weather["Temp3pm"].mean(), inplace = True)
```
